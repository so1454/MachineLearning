##  1. 미스터리 사인 
- 문제와 정답을 준 상태에서 규칙을 찾는것

### 1.1 문제1 
- 22 ? 7 = 37 
- 큰 수 + 두수의 차 = 큰수 + (큰수 - 작은 수)

## 2. 데이터 수집

```python
# 보통 이 부분은 거의 데이터를 밖에서 끌어올 것임(캐글처럼) ->전문가들이 하는 부분임!
list1 = []
for i in range(1,101) : # 첫 번째 수
    for j in range(1,101 ) : # 두 번째 수
        # i j는 문제 데이터
        result = (i + (i-j)) if i > j else (j + (j-i))  # 삼항 연산자
        list1.append([i,j,result])
```

### 2.2 데이터처리

```python
# 데이터를 전처리해줌우리가 보기 쉽게/컴퓨터가 알아보기 쉽게

import pandas as pd
data1 = pd.DataFrame(list1,columns = ['N1','N2','Result'])
data1.head()
```

```python
# 문제(X)와 정답(Y)을 나누자
x = data1.loc[ : , 'N1' : 'N2']
y = data1.loc[ : , 'Result']
```

### 2.3 머신러닝 모델 불러오기

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
pip install scikit-learn
```

### 2.4 머신러닝 모델 학습
- 규칙을 찾는다

```python
rf.fit(x,y) #위에서 담은 xy를 담아서 배울 수 있도록 해줌
```

### 2.5 찾은 규칙으로 예측 
- 대부분의 모델은 규칙을 사람이 이해하기는 힘듬

```python
# 그래서 직접 숫자를 예측치로 넣어서 확인해보면!
# 예측한 수를 출력해줌(위에서 큰 수 + 두수의 차 = 큰수 + (큰수 - 작은 수)를 적용! )

rf.predict([[22, 7]])
```

## 3 문제2
- 4 ? 19 = 3
- (큰 수 / 작은 수) 의 나머지

### 3.1 데이터 수집

```python
list2 = [];
for i in range(1, 101) :
    for j in range(1,101) :
        if i>j :
            result =i%j
        else :
            result = j%i
        list2.append([i,j,result])
list2
```

### 3.2 데이터 처리

```python
data2 = pd.DataFrame(list2, columns= ['M1','M2','Result'])
data2.tail(20)
```

### 3.3 담아주기

```python
x1 = data2.loc[ : , 'M1' : 'M2']
y1 = data2.loc[ : , 'Result']
```

### 3.4 찾은 규칙으로 예측

```python
rf.fit(x1,y1)
```

### 3.5 찾은 규칙으로 예측

```python
rf.predict([[4,19]])
```
