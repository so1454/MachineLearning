```python
import pandas as pd
```

```python
# 딕셔너리로 데이터 프레임 만들기
df = pd.DataFrame(data={
    '진로희망' : ['빅데이터분석가','자영업','웹개발자','대학원','합계'],
    '도수' : [7, 6, 4, 3, 20]
})
```

```python
df
```

```python
df2 = pd.DataFrame(data={
    '진로희망' : ['빅데이터분석가','자영업','웹개발자','대학원','합계'],
    '도수' : [12, 14, 10, 14, 50]
})
```

```python
df2


# ### 빅데이터 반
```

```python
df_concat = pd.concat([df,df2['도수']], axis = 1)
df_concat.columns = ['진로희망','인공지능반','빅데이터반']
df_concat


# ### 상대도수 데이터 프레임 만들기
```

```python
df_list = pd.DataFrame([[6/20, 12/50],
                      [7/20, 14/50],
                      [4/20, 10/20],
                      [3/20, 14/50],
                      [1,1]],
                      columns=['인공지능반','빅데이터반'])
```

```python
df_list


# ### 데이터 프레임 병합
```

```python
df_rs = pd.concat([df_concat, df_list], axis = 1)
df_rs
```

```python
# 멀티 컬럼 작성하기
df_rs.columns = [["", '도수','도수','상대도수','상대도수'],
                ['진로희망','인공지능반','빅데이터반','인공지능반','빅데이터반']]
df_rs
```

```python
# 도수 : (빈도수) 데이터의 개수
# 상대도수 : 도수에서 전체 수를 나눈 것(%)
#            총 도수가 다른 두 데이터를 비교할 때 상대도수를 이용
```
